```python
def get_lat_long(x):
    geolocator = Nominatim(user_agent="myGeocoder")
    location = geolocator.geocode(x)
    try:
        return [location.latitude, location.longitude]

    except: 
        return "unknown"
```

```python
geoloc = {}
for k in monuments:
    geoloc[k] = get_lat_long(k)
```

```python
geoloc
```

```python
latlon = pd.DataFrame.from_dict(geoloc, orient='index')
latlon
```
